# Load dataset

```python
import pandas as pd

df = pd.read_csv("CC_GENERAL.csv")
df
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df['CREDIT_LIMIT'].fillna(df['CREDIT_LIMIT'].mean(), inplace=True)
df['MINIMUM_PAYMENTS'].fillna(df['MINIMUM_PAYMENTS'].mean(), inplace=True)
df.drop(["CUST_ID"], axis=1, inplace=True)
df.isnull().sum().sum()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
def plot_correlation_map(df):
    corr = df.corr()
    s , ax = plt.subplots(figsize = (12 , 10))
    cmap = sns.diverging_palette(220 , 10 , as_cmap = True)
    s = sns.heatmap(
        corr, 
        cmap = cmap,
        square=True, 
        cbar_kws={ 'shrink' : .9 }, 
        ax=ax, 
        annot = True, 
        annot_kws = { 'fontsize' : 12 }
        )
plot_correlation_map(df)
```

# Hierarchical clustering

```python
from sklearn.preprocessing import StandardScaler, normalize

# Standardize data
scaler = StandardScaler() 
scaled_df = scaler.fit_transform(df) 
  
# Normalizing the Data 
normalized_df = normalize(scaled_df) 
 
# Converting the numpy array into a pandas DataFrame 
normalized_df = pd.DataFrame(normalized_df) 
```

```python
from sklearn.cluster import AgglomerativeClustering
model=AgglomerativeClustering(n_clusters=3, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(normalized_df)
agglomerative=pd.DataFrame(clust_labels)
agglomerative
```

# Plot the clusters

```python
fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (normalized_df[13] , normalized_df[2] , c=agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("PAYMENTS")
ax.set_ylabel("PURCHASES")
plt.colorbar(scatter)
```

# Plot the dendrogram

```python
import scipy.cluster.hierarchy as shc
plt.figure(figsize=(10,7))
plt.title("Dendrograms")
dend=shc.dendrogram(shc.linkage(normalized_df, method="complete"))
```

# K-means clustering

```python
from sklearn.cluster import KMeans
kmeans=KMeans(n_clusters=3, random_state=0)
kmeans.fit(normalized_df)
labels=pd.DataFrame(kmeans.labels_)
labels
```

#  Select the best k value using the elbow method

```python
sum_of_squared_distances = []
K = range(1, 15)
for k in K:
    km = KMeans(n_clusters=k)
    km = km.fit(normalized_df)
    sum_of_squared_distances.append(km.inertia_)
    
plt.plot(K, sum_of_squared_distances, 'bx-')
plt.xlabel('k')
plt.ylabel('sum_of_squared_distances')
plt.show()
```

Thanks to the elbow method the optimal k value is 5.

```python
kmeans=KMeans(n_clusters=5, random_state=0)
kmeans.fit(normalized_df)
labels=pd.DataFrame(kmeans.labels_)
labels

plt.scatter(normalized_df[13], normalized_df[2], c = labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title("K-means Clutering")
plt.xlabel("PAYMENTS")
plt.ylabel("PURCHASES")
plt.show()
```

# Compare the two results

```python
# Hierarchical clustering
model=AgglomerativeClustering(n_clusters=5, affinity='euclidean', linkage='complete')
clust_labels=model.fit_predict(normalized_df)
agglomerative=pd.DataFrame(clust_labels)
agglomerative

fig =plt.figure()
ax = fig.add_subplot(111)
scatter = ax.scatter (normalized_df[13] , normalized_df[2] , c=agglomerative[0], s=50)
ax.set_title("Agglomerative Clutering")
ax.set_xlabel("PAYMENTS")
ax.set_ylabel("PURCHASES")
plt.show()

# K-means clustering
kmeans=KMeans(n_clusters=5, random_state=0)
kmeans.fit(normalized_df)
labels=pd.DataFrame(kmeans.labels_)
labels

plt.scatter(normalized_df[13], normalized_df[2], c = labels)
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s=300, c='red', label = 'Centroids')
plt.title("K-means Clutering")
plt.xlabel("PAYMENTS")
plt.ylabel("PURCHASES")
plt.show()
```

The hierarchical clustering and k-means clustering plots are similar. 

# Another validation metric => PCA

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
principal_comp = pca.fit_transform(normalized_df)
pca_df = pd.DataFrame(data=principal_comp,columns=['pca1','pca2'])
labels = kmeans.labels_
pca_df = pd.concat([pca_df,pd.DataFrame({'cluster':labels})], axis = 1)
plt.figure(figsize=(10,10))
plt.style.use('ggplot')
ax = sns.scatterplot(x="pca1", y="pca2", hue = "cluster", data = pca_df, palette =['red','green','blue','black','yellow'])
plt.show()
```
